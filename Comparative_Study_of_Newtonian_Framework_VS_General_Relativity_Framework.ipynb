**Comparative Study of Newtonian Framework vs General Relativity Framework**

Abstract

This project aims to demonstrate the differences between Newtonian physics and General Relativity (GR) in modelling the merger of black hole–neutron star binaries. Using a dataset from Kaggle and the PyCBC software, I generated waveform models to calculate the peak frequencies of these mergers. The Fast Fourier Transform (FFT) method was employed to analyse the frequency components. The results show that the peak frequency predicted by GR is significantly higher than that predicted by Newtonian physics. This finding emphasises that Newtonian approximations are insufficient for accurately describing such complex systems, highlighting the need to use full relativistic models in gravitational wave studies.



Literature Review

Introduction

Gravitational waves are ripples in spacetime, predicted by Albert Einstein in 1916 as a result of his General Theory of Relativity [1], [2]. Although there was some debate among scientists at the time, Einstein himself questioned their existence in a paper titled "Do Gravity Waves Exist?", co-authored with Nathan Rosen. However, he later retracted that view after realizing an error in the paper [3].

Einstein’s theory of General Relativity redefined our understanding of gravity. Instead of being a force, as Newton described, gravity is seen as the result of spacetime being curved by mass and energy [4].

Gravitational waves are produced by nearly all moving objects in the universe, but only the ones generated by the acceleration of very massive objects like binary black holes, neutron stars, or black hole neutron star pairs, that are strong enough to be detected by our instruments [2]. As these objects spiral inwards, they emit gravitational waves and lose energy, with the waves travelling at the speed of light. These waves interact very weakly with matter, which allows them to carry clear information about the merger events that produced them [2].

Gravitational waves are important because they usually come from the most energetic and violent events in the universe, such as colliding black holes and merging neutron stars. Studying them could help us understand the early moments of the universe, which are also believed to have involved such violent processes. This provides a different method to study the universe—complementing electromagnetic observations and possibly revealing phenomena that light-based methods might miss.

Gravitational Wave Detectors

Joseph Weber was the first to start building a detector for measuring gravitational waves. In 1966, he and his team built a device that consisted of an antenna surrounded by a belt of detectors [2]. The antenna was a massive aluminium cylinder. The detectors were placed 950 km apart to make sure local disturbances did not interfere with the readings [2]. The detectors were connected by a high-speed phone line. The published results from Weber’s detector were concluded as unlikely, as they were associated with irregular events.

The discovery of the Hulse-Taylor binary led to the construction of more sensitive and wide-band detectors [1], [5]. Today, there is an entire worldwide network of interferometric detectors for gravitational waves. The network at two sites in the US is called LIGO, which consists of two detectors, each with 4 km long arms [6]. There is also a 3 km arm-length detector called VIRGO in Pisa, Italy, and a 600 m arm-length interferometer called GEO600 in Hannover, Germany. Many new detectors are being developed, which will help provide more accurate results in the future.

Methodology

Motivation

This is a three-part paper. The first two parts are written by William H. Lee and Włodzimierz Kluźniak, while the third part is written by William H. Lee alone.

In the first paper, the value of gamma (i.e., the equation of state of the neutron star) is taken as 3, and the binary system is assumed to be tidally locked [7]. The main aim of the study is to understand the possible outcomes of a black hole–neutron star (BH–NS) merger. Specifically, the authors investigate whether the neutron star would be tidally disrupted, form an accretion disk around the black hole, or directly plunge into it [7].

The focus is placed on how physical parameters such as the mass ratio and the stiffness of the neutron star affect the dynamics of the system and the gravitational wave emission [7]. By using a tidally locked binary configuration, the authors establish a baseline scenario with clearly defined initial conditions. This allows them to isolate the effects of mass ratio and the neutron star’s equation of state [7].

The work is also intended to serve as a comparison with previous simulations of double neutron star mergers conducted by Rasio and Shapiro [8].

Numerical Method

The simulation uses Smoothed Particle Hydrodynamics (SPH), a mesh-free Lagrangian method [7]. In this approach, the fluid is divided into particles, each carrying mass, density, pressure, and velocity. The fluid equations are then solved using interpolated sums of neighbouring particles via a kernel function [7].

Gravitational forces are computed using a tree-based algorithm, which helps speed up calculations for long-range interactions. Time integration is performed using an explicit time-stepping method.

Modelling of the Neutron Star

The black hole–neutron star merger is modelled similarly to neutron star–neutron star or white dwarf mergers. The SPH code is calibrated using results from Rasio and Shapiro’s earlier work.

The neutron star matter is compressible and follows a polytropic equation of state [7]:

Modelling of the Black Hole

The black hole is treated as a Newtonian point mass generating a potential given by [7],

ΦBH(𝑟) = −𝐺𝑀BH/𝑟 [7]

An absorbing boundary is placed at a radius defined by:

𝑟𝑠 = 2𝐺𝑀BH/𝑐^2 [7]

This represents the Schwarzschild radius. If any SPH particle crosses this boundary during the simulation, it is removed [7]. The black hole’s mass, position, and velocity are then updated to ensure conservation of total mass and linear momentum [7].

Result from the 1st paper

Newtonian Hydrodynamics of the Coalescence of Black Holes with Neutron Stars I: Tidally locked binaries with a stiff equation of state.

The conclusions drawn from the simulation by the authors are as follows:

1. If the black hole is much more massive than the neutron star, the star is completely swallowed, with no accretion disk forming. However, if the mass ratio is closer (i.e. the black hole is not overwhelmingly more massive), the neutron star is tidally disrupted, leading to the formation of an accretion disk [7].

2. Tidal disruption causes a sudden suppression of the gravitational wave signal near the moment of merger [7].

3. When the neutron star is swallowed whole, the gravitational wave signal ends more abruptly. In contrast, if a disk forms, the resulting waveform is more complex [7].

In the third part of the paper, the author keeps the same initial conditions, numerical methods, black hole and neutron star masses, and uses the same two gamma values: 3 (stiff) and 5/3 (soft) [10]. The main difference is that the neutron star is no longer tidally locked to the black hole. Instead, it is set up in a more realistic, non-spinning state [10]. This change leads to a sharper and more episodic mass transfer during the merger. The accretion disk that forms around the black hole is more clumpy and irregular compared to earlier cases. Overall, this simulation gives a better picture of what actual neutron star–black hole inspirals would look like in space [10].

Plotting Amplitude vs Frequency to check which gravitational wave frequencies are the strongest.

The input for this function are,
1. Time Array- the time values used in the output files in secs.
2. Signal- either the h_plus polarisation or the h_cross.

dt is the time interval, N is the length of the signal.

fft- applies fast fourier transformation, it decomposes the signal into sine/cosine components at different frequencies.

fftfreq- generates positive and negative frequencies. In this case we have taken only the h_plus value (the real signal has symmetry).

np.hanning- multiples the signal with a Hanning window to a smooth taper. It reduces the spectral leakage caused by signal discontiniutes at the ends.

np.abs- gives amplitude (magnitude) of the complex fft result, fft is a complex number (a+ib).

The output given by the function are,
1. Plot for the positive frequencies.
2. Corresponding amplitudes in the frequency domain.
3. Peak frequency value for each gamma and spin configuration.

```python
def compute_fft(time, signal):
    N = min(len(time), len(signal))
    time = time[:N]
    signal = signal[:N]
    dt = time[1] - time[0]
    freqs = fftfreq(N, dt)
    fft_result = fft(signal * np.hanning(N))
    positive_freqs = freqs[:N // 2]
    amplitude_spectrum = np.abs(fft_result[:N // 2])
    peak_idx = np.argmax(amplitude_spectrum)
    peak_freq = positive_freqs[peak_idx]
    return positive_freqs, amplitude_spectrum, peak_freq

time_signals = [t, t2, t3, t4, t5]
plus_signals = [rh_plus, rh_plus2, rh_plus3, rh_plus4, rh_plus5]
titles = ["Gamma=3.0 (Irrotational)", "Gamma=2.0 (Irrotational)", "Gamma=5/3 (Irrotational)", "Gamma=3.0 (Tidally Locked)", "Gamma=5/3 (Tidally Locked)"]

peak_frequencies = []
for i in range(len(time_signals)):
    freqs, amps, peak_freq = compute_fft(time_signals[i], plus_signals[i])
    peak_frequencies.append(peak_freq)

    plt.figure(figsize=(10, 6))
    plt.plot(freqs, amps)
    plt.axvline(peak_freq, color='red', linestyle='--', label=f'Peak Frequency: {peak_freq:.2f} Hz')
    plt.title(f"FFT of h+ – {titles[i]}")
    plt.xlabel("Frequency (Hz)")
    plt.ylabel("Amplitude (a.u.)")
    plt.grid(True)
    plt.xlim(100, 150)
    plt.legend()
    plt.tight_layout()
    plt.show()

print("Peak Frequencies:")
for i, freq in enumerate(peak_frequencies):
    print(f"{titles[i]}: {freq:.2f} Hz")
```

Interpretation of the Plot:
1. The sharp peak in frequency spectrum shows the dominant gravitational wave frequency emitted during the merger.
2. It usually tells us the orbital frequency of the binary system just before the inspiral or the ringdown merger.
3. When the value of gamma and the spin configuration (irratational and tidal locked) is changed the peak frequency is shifted. This mostly indicates how the internal structure and the spin of the neutron star affects the dynamics of the merger than it's affect on the waveform.
4. The variations in the peak frequencies is quiet less, this would mean that the newtonian dynamics does not strongly capture the impact of internal neutron star structure (Equation of State of the neutron star) or the spin-orbit interactions.
5. Since this particular result uses Newtonian Framework, the peak frequencies are lower than what is expected from the GR framework.  

Relativistic Framework


Plotting of the waveform

```python
m1 = 4.516 #Mass of the Balck Hole in Solar Masses
m2 = 1.4 #Mass of the Neutron star in Solar Masses

from pycbc.waveform import get_td_waveform

hp, hc = get_td_waveform(approximant="SEOBNRv4",
                         mass1=m1,
                         mass2=m2,
                         delta_t=1.0/8192,
                         f_lower=10)

plt.figure(figsize=(15, 5))
plt.plot(hp.sample_times, hp, label='Plus Polarisation', color='blue',linestyle='-')
plt.plot(hc.sample_times, hc, label='Cross Polarisation', color='orange', linestyle='--')
plt.xlabel('Time (s)')
plt.ylabel('Strain')
plt.title('Gravitational Wave Polarizations')
plt.legend()
plt.grid(True)
plt.show()
```

Calculating the Peak Frequency using FFT

```python
time_window = 0.5
times = hp.sample_times.numpy()
strain = hp.numpy()

mask = times > (times[-1] - time_window)
times_merger = times[mask]
strain_merger = strain[mask]

freqs, amp_spec, peak_freq = compute_fft(times_merger, strain_merger)

plt.figure(figsize=(10, 5))
plt.plot(freqs, amp_spec, label="Amplitude Spectrum")
plt.axvline(peak_freq, color='red', linestyle='--', label=f'Peak Frequency: {peak_freq:.2f} Hz')
plt.xlim(50, 300)
plt.xlabel("Frequency (Hz)")
plt.ylabel("Amplitude")
plt.title("FFT of Gravitational Wave Strain Near Merger (10–200 Hz)")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()

print(f"Peak frequency near merger: {peak_freq:.2f} Hz")
```

Conclusion

To summarise with, this project shows that Newtonian physics cannot accurately predict the frequency of gravitational waves at the point of merger. While Newtonian physics works well for the earlier stages, it does not account for the complex effects near the merger. The frequency calculated using General Relativity is higher, which reflects the true behaviour of the system in strong gravity. This means that to understand gravitational wave signals properly, especially during merger, we need to use the full theory of General Relativity rather than Newtonian approximations.
